### **KPI's**

```python
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
total_orders = df['Order ID'].nunique()
aov = total_sales / total_orders
total_customers = df['Customer Name'].nunique()
avg_profit_order = total_profit / total_orders
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
avg_discount = df['Discount'].mean() * 100
profit_margin = (total_profit / total_sales) * 100
best_subcat = df.groupby('Sub Category')['Sales'].sum().sort_values(ascending=False).head(5)
worst_subcat = df.groupby('Sub Category')['Sales'].sum().sort_values().head(5)
```

```python
print("=========== SUPER MART GROCERY SALES ‚Äì KPIs ===========")

print(f"üìå Total Sales: ‚Çπ{total_sales:,.2f}")
print(f"üìå Total Profit: ‚Çπ{total_profit:,.2f}")
print(f"üìå Total Orders: {total_orders:,}")
print(f"üìå Total Customers: {total_customers:,}")

print("\n=========== AVERAGES ===========")
print(f"üìå Average Order Value (AOV): ‚Çπ{aov:,.2f}")
print(f"üìå Avg Profit per Order: ‚Çπ{avg_profit_order:,.2f}")
print(f"üìå Average Discount Given: {avg_discount:.2f}%")
print(f"üìå Profit Margin: {profit_margin:.2f}%")

print("\n=========== CATEGORY PERFORMANCE ===========")
print("üì¶ Sales by Category:")
print(category_sales)

print("\nüì¶ Profit by Category:")
print(category_profit)

print("\n=========== SUB-CATEGORY PERFORMANCE ===========")
print("‚≠ê Top 5 Best Selling Sub-Categories:")
print(best_subcat)

print("\nüîª Bottom 5 Least Selling Sub-Categories:")
print(worst_subcat)

print("\n=========== REGION PERFORMANCE ===========")
print("üåç Sales by Region:")
print(region_sales)

print("\n======================================================")


```

### **Chart's**

#### **Monthly Sales Trend**

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

#### **Sales by Category**

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values()

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color='steelblue')
plt.title("Sales by Category")
plt.ylabel("Total Sales")
plt.xlabel("Category")
plt.show()
```

#### **Profit by Category**

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values()

plt.figure(figsize=(8,5))
category_profit.plot(kind='bar', color='purple')
plt.title("Profit by Category")
plt.ylabel("Total Profit")
plt.xlabel("Category")
plt.show()
```

#### **Sales by Region**

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values()

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar', color='green')
plt.title("Sales by Region")
plt.ylabel("Total Sales")
plt.xlabel("Region")
plt.show()
```

#### **Top 5 Sub-Categories (Sales)**

```python
top_sub = df.groupby('Sub Category')['Sales'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top_sub.plot(kind='barh', color='orange')
plt.title("Top 5 Sub-Categories by Sales")
plt.xlabel("Sales")
plt.ylabel("Sub Category")
plt.gca().invert_yaxis()
plt.show()
```

#### **Bottom 5 Sub-Categories (Sales)**

```python
bottom_sub = df.groupby('Sub Category')['Sales'].sum().sort_values().head(5)

plt.figure(figsize=(8,5))
bottom_sub.plot(kind='barh', color='red')
plt.title("Bottom 5 Sub-Categories by Sales")
plt.xlabel("Sales")
plt.ylabel("Sub Category")
plt.show()
```

#### **Discount vs Profit**

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Discount'], df['Profit'], alpha=0.4)
plt.title("Discount vs Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.show()
```

#### **Sales Distribution**

```python
plt.figure(figsize=(8,5))
plt.hist(df['Sales'], bins=20, color='skyblue')
plt.title("Distribution of Sales")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```
